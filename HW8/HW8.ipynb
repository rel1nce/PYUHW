# Task1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = np.array([[5, 3], [10, 15], [15, 12], [24, 10], [30, 45], [85, 70], [71, 80], [60, 78], [55, 52], [80, 91]])


def clusterize(n):
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(x)
    y_kmeans = kmeans.predict(x)

    plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=20, cmap='summer')
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='blue', s=100, alpha=0.9)


plt.figure(figsize=(12.8, 4))

plt.subplot(131)
clusterize(2)
plt.subplot(132)
clusterize(3)
plt.subplot(133)
clusterize(4)

plt.show()
```

# Task2

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

iris = load_iris()

kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(iris.data)

_, ax = plt.subplots(1, 2, figsize=(9, 4))
centers = kmeans.cluster_centers_.reshape(3, 4)

ax[0].scatter(iris.data[:, 0], iris.data[:, 1], c=clusters, s=20, cmap='summer')
ax[0].scatter(centers[:, 0], centers[:, 1], c='blue', s=100, alpha=0.9)
ax[0].set_xlabel('Sepal Length (cm)')
ax[0].set_ylabel('Sepal Width (cm)')

ax[1].scatter(iris.data[:, 2], iris.data[:, 3], c=clusters, s=20, cmap='summer')
ax[1].scatter(centers[:, 2], centers[:, 3], c='blue', s=100, alpha=0.9)
ax[0].set_xlabel('Petal Length (cm)')
ax[0].set_ylabel('Petal Width (cm)')

plt.show()
```

# Task3

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import load_iris

iris = load_iris().data[:, :2]

linked = linkage(iris, 'ward')

plt.figure(figsize=(10, 5))
dendrogram(linked, distance_sort='descending')
plt.show()

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
cluster.fit_predict(iris)

plt.figure(figsize=(4.2, 4))
plt.scatter(iris[:, 0], iris[:, 1], c=cluster.labels_, cmap='rainbow')
plt.show()
```
