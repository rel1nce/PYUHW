```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

figure = plt.figure()
axes = figure.add_subplot(111, projection = '3d')

point1 = np.array([0, 0, 0])
point2 = np.array([1, 1, 1])
point3 = np.array([2, 2, 2])
point4 = np.array([3, 3, 3])

axes.scatter(0, 0, 0)
axes.scatter(1, 1, 1)
axes.scatter(2, 2, 2)
axes.scatter(3, 3, 3)

print(np.linalg.norm(point1 - point2))
print(np.linalg.norm(point1 - point2) ** 2)
print(np.linalg.norm(point1 - point2, ord = np.inf))
print(np.linalg.norm(point1 - point2, ord = 1))

print(np.linalg.norm(point1 - point3))
print(np.linalg.norm(point1 - point3) ** 2)
print(np.linalg.norm(point1 - point3, ord = np.inf))
print(np.linalg.norm(point1 - point3, ord = 1))

print(np.linalg.norm(point1 - point4))
print(np.linalg.norm(point1 - point4) ** 2)
print(np.linalg.norm(point1 - point4, ord = np.inf))
print(np.linalg.norm(point1 - point4, ord = 1))

plt.show()
```

# 2.3.1

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# data
x = np.array([[-1, 1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]
iris = sns.load_dataset('iris')
iris


x_train, x_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.15
)

x_train.shape, x_test.shape, y_train.shape, y_test.shape

x_train.head()
y_train.head()

k = 5
model = KNeighborsClassifier(n_neighbors = k)
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
y_pred

fig = plt.figure()
sns.scatterplot(x = 'petal_width', y = 'petal_length', 
                data = iris, 
                hue = 'species', 
                s = 50
)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend(loc = 2)
plt.grid()

for ind in range(len(y_test)):
  if np.array(y_test)[ind] != y_pred[ind]:
    plt.scatter(x_test.iloc[ind, 3], x_test.iloc[ind, 2], color = 'red', s = 150)
print(f'accuracy: {accuracy_score(y_test, y_pred) : .3}')
```
