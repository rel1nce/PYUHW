# Task 1

```python
import math as m

class TrigoFunctions:
    def __init__(self):
        pass
    
    def sin(rd):
        return m.sin(rd)
    
    def cos(rd):
        return m.cos(rd)

    def tan(rd):
        return m.tan(rd)
    
    def asin(deg):
        return m.asin(deg)

    def acos(deg):
        return m.acos(deg)

    def atan(deg):
        return m.atan(deg)

    def degToRad(deg):
        return m.radians(deg)


example = TrigoFunctions

print(example.sin(m.pi))
print(example.cos(m.pi))
print(example.tan(m.pi))

print(example.asin(1))
print(example.acos(1))
print(example.atan(1))

print(example.degToRad(45))
```

# Task 3

```python
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def PrintTree(self):
        print(self.data)

    def printTree(self, level = 0):
        if self.left != None:
            self.left.printTree(level + 1)

        print(self.data)

        if self.right != None:
            self.right.printTree(level + 1)

    def add(self, data):
        if data < self.data:
            if self.left == None:
                self.left = Tree(data)
            else:
                self.left.add(data)
        elif data > self.data:
            if self.right == None:
                self.right = Tree(data)
            else:
                self.right.add(data)


binaryTree = Tree(8)

binaryTree.add(5)
binaryTree.add(10)
binaryTree.add(4)
binaryTree.add(1)
binaryTree.add(3)
binaryTree.add(9)
binaryTree.add(2)
binaryTree.printTree()
```

# Task 4

```python
import numpy as np
from sklearn import tree
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
target = [0, 0, 0, 1, 1, 1]

x_train, x_test, y_train, y_test = train_test_split(
    x,
    target,
    test_size=0.20
)

classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)
tree.plot_tree(classifier)

y_pred = classifier.predict(x_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Task 5

```python
import numpy as np
import pandas as pd
from sklearn import tree, metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

url = r'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataset = pd.read_csv(url)

print(dataset.describe())

x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)

tree.plot_tree(regressor)

y_pred = regressor.predict(x_test)

df = pd.DataFrame({'Actual': y_test, 'Prediction': y_pred})
print(df)

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print(metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100)
```
